```python
# TTA
class TestDataset(Dataset):
    def __init__(self, aug_num, test_imgs, transforms=None):
        self.aug_num = aug_num
        self.test_imgs = test_imgs
        self.transforms = transforms

    def __len__(self):
        return len(self.test_imgs)

    def __getitem__(self, index):
        img_path = self.test_imgs[index]
        img_code = (Path(img_path)).parent.stem
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        imgs = []
        for transform_list in self.transforms.values():
            img_list = []
            for transform in transform_list:
                img_list.append(transform(image=img)['image'])
            img_list = np.concatenate(img_list, axis=0)
            img_list = rearrange(
                img_list, '(new1 c) h w -> new1 c h w', new1=self.aug_num)
            imgs.append(img_list)

        

        return imgs, img_code


testDataset = TestDataset(4, test_jpg, transforms=data_transforms_valid)
testDataloader = DataLoader(
    testDataset, batch_size=CONFIG['valid_batch_size'], shuffle=False)
```

# Inference

```python
# no TTA
ic.disable()
with torch.no_grad():

    for model_list in models.values():
        for model in model_list:
            model.eval()

    img_code_list = []
    outputs = []

    bar = tqdm(enumerate(testDataloader), total=len(testDataloader))

    for step, (img_list, img_code) in bar:
        # img_list : list of (bs, N, C, H, W)
        # img_code : (bs)

        probs_list = []
        # element :  (bs, N, num_classes)
        # length : 모델의 개수

        # 각 모델별로 predict한 결과를 모은다
        for i, model_list in enumerate(models.values()):
            img = img_list[i]
            # (bs, N, C, H, W)
            batch_size = img.shape[0]
            img = rearrange(img, 'bs N C H W -> (bs N) C H W')
            img = img.to(device)

            for model in model_list:
                logits = model(img)
                logits = logits.cpu()
                # ( (bs N), num_classes)
                logits = rearrange(logits, '(bs N) num_classes -> bs N num_classes', bs = batch_size)
                # 모델들마다 logit들의 값의 크기가 다를 수 있으므로, softmax를 이용해 정규화하는 작업이 필요하다
                prob = torch.softmax(logits, dim=-1)
                # (bs, N, num_classes)
                probs_list.append(prob)

        probs = torch.stack(probs_list)
        # (num_model, bs, N, num_clases)
        probs = torch.sum(probs, dim=0)
        # (bs, N, num_classes)
        probs = torch.sum(probs, dim=1)
        # (bs, num_classes)


        output = np.argmax(torch.softmax(probs, dim=-1),
                           axis=-1)  # torch.Tensor, (bs,)
        output = np.array(output)
        output = [class_dict[i] for i in output]

        img_code_list.extend(img_code)
        outputs.extend(output)

        assert len(img_code_list) == len(outputs)
```
